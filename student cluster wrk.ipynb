```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv(r'C:\Users\HP\Desktop\students-performance-in-exams\StudentsPerformance.csv')
data
```

```python
data_mapped  = data.copy()
data_mapped['gender']= data_mapped['gender'].map({'female':0,'male':1})
data_mapped
```

```python
x= data_mapped.iloc[:,0:7]
x
```

```python
plt.scatter(data['math score'],data['reading score'])
plt.xlim()
plt.ylim()
plt.show()
```

```python
x = data.iloc[:,5:]
x
```

```python
kmeans= KMeans(3)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters= data.copy()
data_with_clusters['clusters'] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['math score'],data_with_clusters['reading score'],c=data_with_clusters['clusters'],cmap='rainbow')
plt.xlim()
plt.ylim()
plt.show()
```

```python
wcss=[]

for i in range (1,7):
    kmeans=KMeans(i)
    kmeans.fit(x)
    wcss_iter=kmeans.inertia_
    wcss.append(wcss_iter)
    
wcss
    
```

```python
number_clusters=range(1,7)
plt.plot(number_clusters,wcss)
plt.xlabel('number_clusters')
plt.ylabel('within=clusters sum of squares')
```
